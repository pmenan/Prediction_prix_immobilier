# Choix du modèle et Réalisation des prédictions

```python
#isolation de la variable cible du jeu de données de départ
y = data_set['median_house_value']
X = data_set.drop('median_house_value', axis = 1)
```

```python
#vérification des valeurs de X et y
X.head()
```

```python
y.head()
```

```python
#séparation du data_set en deux sous ensemble (apprentissage et test)
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
#vérification de la séparation des données en sous ensemble
X_train.head()
```

```python
X_valid.head()
```

```python
y_train.head()
```

```python
y_valid
```

```python
#transformation des données avant définition de modèle.

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

#données numérique : faire de l'imputation et rendre les donnée standard
num_transformer = Pipeline(steps = [
    ('impute', SimpleImputer()),
    ('scaler', StandardScaler())
])

#données catégorielles (une seule colonne dans ce cas de figure) : faire de l'encodage en une seule fois.
categorical_transformer = OneHotEncoder(handle_unknown = 'ignore', sparse = False)

#colonnes catégorielle à considérer pour le traitement des données 
cat_column = ['ocean_proximity']

#colonnes numériques à considérer pour le traitement des données
num_column = ['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households',
            'median_income']

#étapes de prétraitement des données en une seule fois
preprocessor = ColumnTransformer(transformers = [
    ('num_trans', num_transformer, num_column),
    ('categorical_trans', categorical_transformer, cat_column)
])
```

```python
"""
Normalement un modèle de regression linaire simple pourrait être adapter pour résoudre ce problème.
mais pour des raisons de cherche de meilleures performance, plusiuers modèles seront testés 

"""

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score

#définition de modèle.
linear_model = LinearRegression()
tree_model = DecisionTreeRegressor(random_state =0)
forest_model = RandomForestRegressor(random_state = 0)
sgd_model = SGDRegressor()

model_dict = {
    'line_model' : linear_model,
    'tree_model' : tree_model,
    'forest_model' : forest_model,
    'sgd_model' : sgd_model 
}

print("score sans cross validation\n")
for name, model in model_dict.items(): 
    my_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor), 
        ('model', model)
])
    my_pipeline.fit(X_train, y_train)
    print("le score du modèle ",name, "sur le jeu de validation est de : ", my_pipeline.score(X_valid, y_valid))

print("\n")
print("score avec cross validation\n")
for name, model in model_dict.items(): 
    my_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor), 
        ('model', model)
])
    my_pipeline.fit(X_train, y_train)
    score = -1*cross_val_score(my_pipeline, X_valid, y_valid, cv = 5, scoring = 'neg_mean_absolute_error')
    print("le score en cross validation du modèle ",name, "sur le jeu de validation est de : ", score.mean())

```

```python
#d'après les différents résultats, le modèle de fôret aléatoire semble être le meilleur modèle pour nos données.
#optimisation du paramètre n_estimator notre modèle 
n_estimators = [100, 200, 300, 400, 500, 700, 800, 1000]
for n in n_estimators : 
    model = RandomForestRegressor(n_estimators = n, random_state = 0)
    my_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor), 
        ('model', model)
])
    my_pipeline.fit(X_train, y_train)   
    print('le score pour', n, ' est : ',  my_pipeline.score(X_valid, y_valid))
```

```python
#la valeur n_estimators = 1000 semble être celle offrant un meillleur score.
#prédiction
prediction = my_pipeline.predict(X_valid)
```

```python
#valeurs des prédiction
prediction
```

```python
#erreur de generalisation 
mean_absolute_error(y_valid, prediction)
```
